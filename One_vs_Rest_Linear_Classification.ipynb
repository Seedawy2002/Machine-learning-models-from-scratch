## Linear Classification


### Split dataset

```python
#Randomize data
setosa_sample = irisdf.iloc[:50,:].sample(frac = 1)
versicolor_sample =  irisdf.iloc[50:100,:].sample(frac = 1)
viriginica_sample =  irisdf.iloc[100:150,:].sample(frac = 1)
irisdf_train_concat = pd.concat([setosa_sample, versicolor_sample, viriginica_sample], axis = 0)
irisdf_train = irisdf_train_concat.reset_index(drop=True)
irisdf_train
```

### Make Z for all feature


```python
#Get test data
setosa_test = irisdf_train.iloc[40:50,:]
versicolor_test =  irisdf_train.iloc[90:100,:]
viriginica_test =  irisdf_train.iloc[140:150,:]
irisdf_test_concat = pd.concat([setosa_test, versicolor_test, viriginica_test], axis = 0)
irisdf_test = irisdf_test_concat.copy(deep=True)
print(irisdf_test)

#Get b & Z
b = irisdf_test.loc[:,'Type'].values
print(b)
irisdf_test.drop(['Type'], axis=1,inplace=True)
z=irisdf_test.values
print(z)
```

###Calculate w
## w=(Z^T * Z)^-1 * Z^T * b


### One Vs All 
Build 3 model


##w1, w2 , w3



```python
setosa_train = irisdf_train.iloc[:40,:]
versicolor_train =  irisdf_train.iloc[50:90,:]
viriginica_train =  irisdf_train.iloc[100:140,:]
```

```python
#W1
setosa = pd.concat([setosa_train, versicolor_train, viriginica_train], axis=0)
matrix_setosa = setosa.loc[:,'Type'].replace(1,-1).replace(2,-1).replace(0,1)
setosa.drop(['Type'], axis = 1, inplace = True)
setosa = pd.concat([setosa, matrix_setosa], axis = 1).reset_index(drop=True)
setosa.loc[40:,:] = -1*setosa.loc[40:,:]
setosa_b = setosa.loc[:,'Type'].values
setosa.drop(['Type'], axis=1,inplace=True)
setosa_z=setosa.values
print(setosa_b)
print(setosa_z)
```

```python
#W2
versicolor = pd.concat([versicolor_train, viriginica_train, setosa_train], axis=0)
matrix_versicolor = versicolor.loc[:,'Type'].replace(0,-1).replace(2,-1)
versicolor.drop(['Type'], axis = 1, inplace = True)
versicolor = pd.concat([versicolor, matrix_versicolor], axis = 1).reset_index(drop=True)
versicolor.loc[40:,:] = -1*versicolor.loc[40:,:]
versicolor_b = versicolor.loc[:,'Type'].values
versicolor.drop(['Type'], axis=1,inplace=True)
versicolor_z=versicolor.values
print(versicolor_b)
print(versicolor_z)
```

```python
#W3
viriginica = pd.concat([viriginica_train, setosa_train, viriginica_train], axis=0)
matrix_viriginica = viriginica.loc[:,'Type'].replace(0,-1).replace(1,-1).replace(2,1)
viriginica.drop(['Type'], axis = 1, inplace = True)
viriginica = pd.concat([viriginica, matrix_viriginica], axis = 1).reset_index(drop=True)
viriginica.loc[40:,:] = -1*viriginica.loc[40:,:]
viriginica_b = viriginica.loc[:,'Type'].values
viriginica.drop(['Type'], axis=1,inplace=True)
viriginica_z=viriginica.values
print(viriginica_b)
print(viriginica_z)
```

Evaluation

Class #1
if w1^T * x  > 0
and w2^T * x < 0
and w3^T * x < 0

Class #2
if w1^T * x  < 0
and w2^T * x > 0
and w3^T * x < 0

Class #3
if w1^T * x  < 0
and w2^T * x < 0
and w3^T * x > 0

Undefine Class
if
 w1^T * x    > 0 and w2^T * x > 0 and w3^T * x < 0

OR

w1^T * x    < 0 and w2^T * x > 0 and w3^T * x > 0

OR

w1^T * x    > 0 and w2^T * x < 0 and w3^T * x > 0


```python
#setosa
w_setosa =np.linalg.inv(setosa_z.T@setosa_z)@setosa_z.T@setosa_b
T_setosa = z@w_setosa
T_setosa[T_setosa >= 0] = 1
T_setosa[T_setosa < 0] = -1
print(T_setosa)

#versicolor
w_versicolor =np.linalg.inv(versicolor_z.T@versicolor_z)@versicolor_z.T@versicolor_b
T_versicolor = z@w_versicolor
T_versicolor[T_versicolor >= 0] = 1
T_versicolor[T_versicolor < 0] = -1
print(T_versicolor)

#viriginica
w_viriginica =np.linalg.inv(viriginica_z.T@viriginica_z)@viriginica_z.T@viriginica_b
T_viriginica = z@w_viriginica
T_viriginica[T_viriginica >= 0] = 1
T_viriginica[T_viriginica < 0] = -1
print(T_viriginica)
```

```python
#Get the result
length = len(T_viriginica)
result = []
for i in range(length):
  if T_setosa[i] == 1 and T_versicolor[i] == -1 and T_viriginica[i] == -1:
    result.append('setosa')
  elif T_setosa[i] == -1 and T_versicolor[i] == 1 and T_viriginica[i] == -1:
    result.append('versicolor')
  elif T_setosa[i] == -1 and T_versicolor[i] == -1 and T_viriginica[i] == 1:
    result.append('viriginica')
  elif T_setosa[i] == -1 and T_versicolor[i] == -1 and T_viriginica[i] == -1:
    result.append('None')
  else:
    result.append('Undefined')
result
```

```python
#Analysis
true_counter = 0
for i in range(length):
  if i>=0 and i<10 and result[i] == 'setosa':
    true_counter += 1
  if i>=10 and i<20 and result[i] == 'versicolor':
    true_counter += 1
  if i>=20 and i<30 and result[i] == 'viriginica':
    true_counter += 1
Accuracy = true_counter/length*100
print("Accuracy:", Accuracy, "%")
```
